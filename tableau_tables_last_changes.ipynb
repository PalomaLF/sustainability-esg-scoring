```python
import pandas as pd
import numpy as np
```

```python
country_codes = pd.read_excel('data/tables tableau/country_codes.xlsx')
```

```python
country_global_score = pd.read_excel('data/tables tableau/country_gobal_score.xlsx')
```

```python
country_goals_score = pd.read_excel('data/tables tableau/goals_score.xlsx')
```

```python
country_global_score.rename(columns={'Country Code ISO3':'ISO (3)'}, inplace = True)
```

```python
country_goals_score.rename(columns = {'Country Code ISO3':'ISO (3)'}, inplace = True)
```

```python
country_global_score
```

```python
country_global_score = pd.merge(country_global_score,country_codes, on=['ISO (3)', 'Country', 'Continent','Region'], how= 'left')
```

```python
country_global_score.to_excel('data/tables tableau/country_gobal_score.xlsx', index = False)
```

```python
country_goals_score = pd.merge(country_goals_score,country_codes, on=['ISO (3)', 'Country', 'Continent', 'Region'], how= 'left')
```

```python
country_goals_score.columns
```

```python
country_goals_score.head(1)
```

```python
country_goals_score = country_goals_score.melt(id_vars=['Country', 'ISO (3)', 'population', 'year', 'Continent', 'Region'], var_name='Goal', value_name='score')
```

```python
country_goals_score.head(15)
```

```python
country_goals_score.to_excel('data/tables tableau/goals_score.xlsx', index = False)
```
